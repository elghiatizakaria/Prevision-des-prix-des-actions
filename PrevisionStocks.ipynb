```python
ticker=["SPY","AAPL","KO","DIS","MSFT"]
```

```python
stocks1=yf.download(ticker,start="2013-01-01",end="2023-01-01")
```

```python
stocks1
```

```python
close=stocks1.loc[:,"Close"].copy()
```

```python
normClose=close.div(close.iloc[0]).mul(100)
normClose
```

```python
normClose.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret=close.pct_change().dropna()
ret.head()
```

```python
ret.describe().T
```

```python
summary=ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(9,5),s=50,fontsize=12)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.02,summary.loc[i,"mean"]+0.02),size=15)
plt.xlabel("Anuual Risk(std)",size=12)
plt.ylabel("Anuual return",size=12)
plt.title("Risk/Return",size=12)
```
