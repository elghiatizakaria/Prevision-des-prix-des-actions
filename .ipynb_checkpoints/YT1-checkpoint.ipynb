```python
microsoft=close.MSFT.copy().to_frame()
```

```python
microsoft
```

```python
microsoft.shift(periods=1)
```

```python
microsoft["lag1"]=microsoft.shift(periods=1)
```

```python
microsoft
```

```python
microsoft["Diff"]=microsoft.MSFT.sub(microsoft.lag1)
```

```python
microsoft
```

```python
microsoft["% change"]=microsoft.MSFT.div(microsoft.lag1).sub(1).mul(100)
microsoft
```

```python
microsoft["Diff2"]=microsoft.MSFT.diff(periods=1)
microsoft
```

```python
microsoft["% change 2"]=microsoft.MSFT.pct_change(periods=1).mul(100)
microsoft
```

```python
del microsoft["Diff2"]
```

```python
microsoft
```

```python
microsoft.rename(columns= {'% change':'Change'}, inplace=True)
```

```python
microsoft
```

```python
del microsoft['lag1']
del microsoft['Diff']
del microsoft['Change']
del microsoft['% change 2']
```

```python
microsoft
```

```python
ret = microsoft.pct_change().dropna()
```

```python
ret
```

```python
ret.plot(kind="hist",figsize=(12,8),bins=100)
plt.show()
```

```python
daily_mean_ret=ret.mean()
daily_mean_ret
```

```python
var_daily=ret.var()
var_daily
```

```python
std_daily=np.sqrt(var_daily)
std_daily
```

```python
ret.std()
```

```python
annual_mean_ret=daily_mean_ret*252
annual_mean_ret
```

```python
annual_var_ret=var_daily*252
annual_var_ret
```

```python
annual_std_ret=np.sqrt(annual_var_ret)
annual_std_ret
```
